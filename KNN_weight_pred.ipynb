```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as wr
wr.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/weight-height.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.Gender = le.fit_transform(df[['Gender']])
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
x= df[['Gender','Height']]
y= df['Weight']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred_train = lr.predict(X_train)
y_pred_train
```

```python
y_pred_test = lr.predict(X_test)
y_pred_test
```

```python
train_accuracy = lr.score(X_train,y_train)
train_accuracy
```

```python
test_accuracy = lr.score(X_test,y_test)
test_accuracy
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
lr_mse_test = mean_squared_error(y_test,y_pred_test)
lr_mse_test
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
kn = KNeighborsRegressor(n_neighbors=5)
```

```python
kn.fit(X_train,y_train)
```

```python
kn_y_pred_train = kn.predict(X_train)
kn_y_pred_train
```

```python
kn_y_pred_test = kn.predict(X_test)
kn_y_pred_test
```

```python
kn_accuracy_train = kn.score(X_train,y_train)
kn_accuracy_train
```

```python
kn_accuracy_test = kn.score(X_test,y_test)
kn_accuracy_test
```

```python
kn_mse_test = mean_squared_error(y_test,kn_y_pred_test)
kn_mse_test
```

```python
compare_model = pd.DataFrame({'Model':['Linear Regression','KNN'],'Train Accuracy':[train_accuracy,kn_accuracy_train],'Test Accuracy':[test_accuracy,kn_accuracy_test],'Test MSE':[lr_mse_test,kn_mse_test]})
compare_model
```
